### Matplotlib 인터페이스

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,9,7]
```

```python
plt.plot(x,y)
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
plt.plot(x, np.sin(x))  # 유니버셜 함수 np.sin(x)  # 상태 기반 인터페이스
```

```python
ax = plt.axes()  # axes => 축의 복수
ax.plot(x, np.sin(x)) # 객체지향 인터페이스
```

### 라인 플롯

#### color

```python
plt.plot(x, np.sin(x), color = 'red')
```

```python
plt.plot(x, np.sin(x), color = 'red')
plt.plot(x, np.cos(x), color = 'g')
```

#### linestyle

```python
plt.plot(x, np.sin(x), color = 'red')
plt.plot(x, np.cos(x), color = 'g', linestyle = '--')
```

#### title & label

```python
plt.plot(x, np.sin(x), color = 'red')
plt.plot(x, np.cos(x), color = 'g', linestyle = '--')
plt.title('Sin & Cos curves')
plt.xlabel('x')
plt.ylabel('f(x)')
```

#### legend

```python
plt.plot(x, np.sin(x), color = 'red', label='sin(x)')
plt.plot(x, np.cos(x), color = 'g', linestyle = '--', label='cos(x)')
plt.title('Sin & Cos curves')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
```

#### lim

```python
plt.plot(x, np.sin(x), color = 'red', label='sin(x)')
plt.plot(x, np.cos(x), color = 'g', linestyle = '--', label='cos(x)')
plt.title('Sin & Cos curves')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim((-1.2, 1.2))
```

#### tight_layout

```python
plt.plot(x, np.sin(x), color = 'red', label='sin(x)')
plt.plot(x, np.cos(x), color = 'g', linestyle = '--', label='cos(x)')
plt.title('Sin & Cos curves')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim((-1.2, 1.2))
plt.tight_layout()
```

### 히스토그램

```python
np.random.seed(1)
z = np.random.randn(1000)  # 표준 정규 분포 난수
```

```python
plt.hist(z)
```

```python
plt.hist(z, bins = 20)
```

```python
x1 = np.random.normal(0, 0.8, 1000)  #평균, 표준편차, 갯수
x2 = np.random.normal(-2, 1.2, 1000)
```

```python
plt.hist(x1)
plt.hist(x2)
```

#### alpha =  투명도

```python
plt.hist(x1, alpha = 0.7)
plt.hist(x2, alpha = 0.3)
```

```python
plt.hist(x1, alpha = 0.7, bins = 20)
plt.hist(x2, alpha = 0.3, bins = 20)
```
